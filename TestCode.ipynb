```python
import math
```

```python
def calculate_hypotenuse(a, b):
  a_squared = a ** 2
  b_squared = b ** 2
  c_squared = a_squared + b_squared
  c = math.sqrt(c_squared)
  return c

a = int(input("Enter side a: "))
b = int(input("Enter side b: "))

print(calculate_hypotenuse(a, b))
```
