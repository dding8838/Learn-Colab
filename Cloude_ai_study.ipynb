```python
from decimal import Decimal, getcontext

# 1. 원하는 정밀도를 설정합니다. (기본값은 28자리)
getcontext().prec = 1000 # 50자리 정밀도로 설정

# 2. Decimal 객체를 생성하여 연산합니다.
# 문자열로 숫자를 전달해야 정확한 십진수 값을 저장할 수 있습니다.
a = Decimal('0.1')
b = Decimal('1.1')
c = Decimal('1.2')

print(a+b == c)
```

```python
#도전과제

score = []

for i in range(0,5):
  num = int(input("점수를 입력해주세요 : "))
  score.append(num)
score.sort(reverse=True)
print(f"최고점은 {score[0]}점 입니다.")
print(f"최저점은 {score[-1]}점 입니다.")
```
