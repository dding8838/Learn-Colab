```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])   # shape (2,2)

x = np.array([5, 6])     # shape (2,)
y = A.dot(x)             # 또는 A @ x
print(y)                 # [1*5+2*6, 3*5+4*6] -> [17, 39]

# 전치와 형태 확인
print(A.shape, x.shape, A.T.shape)
```
