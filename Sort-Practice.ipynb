```python
#버블정렬 알고리즘
sort = [42, 17, 95, 23, 60, 8, 31, 74]

temp = 0
length = len(sort)

for i in range(length-1,0,-1):
  for j in range(0,i,1):
    if sort[j] > sort [j+1]:
      temp = sort[j]
      sort[j] = sort[j+1]
      sort[j+1] = temp

print(sort)
```

```python
#선택정렬 알고리즘
sort = [42, 17, 95, 23, 60, 8, 31, 74]

temp = 0
length = len(sort)

for i in range(0,length-1):
    min = i
    for j in range(i+1,length):
        if sort[min] > sort[j]:
            min = j
    temp = sort[i]
    sort[i] = sort[min]
    sort[min] = temp

print(sort)
```

```python
#삽입정렬
sort = [37, 12, 28, 46, 5, 19]

temp = 0
for i in range(1,len(sort)):
  if sort[i-1] > sort[i]:
    temp = sort[i]
    for j in range(i-1,-2,-1):
      if sort[j] > temp:
        sort[j+1] = sort[j]
      else:
        break
    sort[j+1] = temp

print(sort)
```
