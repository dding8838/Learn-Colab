```python
#DotProduct 점곱 계산

a = [1,2,3]
b = [4,5,6]
answer = 0

for i in range(0,len(a)):
  answer += (a[i]*b[i])
print(answer)
```

```python
#Matrox Multiplication 행렬곱 계산

A = [
    [1, 2],
    [3, 4]
]

B = [
    [5, 6],
    [7, 8]
]

C=[]

total = 0

for i in range(0,len(A)):
  row = []
  for j in range(0,len(A[0])):
    total = 0
    for k in range(0,len(A[0])):
      total += A[i][k]*B[k][j]
    row.append(total)
  C.append(row)

print(C)
```
