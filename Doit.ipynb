```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig

# 1. 4-bit ì–‘ìí™” ì„¤ì •
nf4_config = BitsAndBytesConfig(
   load_in_4bit=True,
   bnb_4bit_quant_type="nf4",
   bnb_4bit_use_double_quant=True,
   bnb_4bit_compute_dtype=torch.bfloat16
)

# 2. Llama 3 8B Instruct ëª¨ë¸ ë¡œë“œ
# ì¶”í›„ ì´ ëª¨ë¸ ê²½ë¡œ(meta-llama/Meta-Llama-3-8B-Instruct)ë¥¼ í•œêµ­ì–´ íŒŒì¸íŠœë‹ ëª¨ë¸ë¡œ ë³€ê²½í•˜ë©´ ë©ë‹ˆë‹¤.
model_id = "meta-llama/Meta-Llama-3-8B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=nf4_config,
    device_map="auto"
)

print("âœ… Llama 3 8B Instruct ëª¨ë¸ì´ Colab GPUì— ì„±ê³µì ìœ¼ë¡œ ë¡œë“œë˜ì—ˆìŠµë‹ˆë‹¤.")
```

```python
import fitz # PyMuPDF ë¼ì´ë¸ŒëŸ¬ë¦¬ (fitzìœ¼ë¡œ import ë©ë‹ˆë‹¤)
import os

# âš ï¸ íŒŒì¼ ê²½ë¡œ ì„¤ì •
# Colab ì™¼ìª½ì— ìˆëŠ” íŒŒì¼ íƒ­ì„ ëˆŒëŸ¬ ì„±ê²½ PDF íŒŒì¼ì„ ì—…ë¡œë“œí•˜ì„¸ìš”.
# íŒŒì¼ ì´ë¦„ì´ 'bible.pdf'ë¼ê³  ê°€ì •í•©ë‹ˆë‹¤.
pdf_path = "/content/drive/MyDrive/Learn_Colabotary/Learn-Colab/kjv.pdf"
extracted_text_path = "bible_full_text.txt"

# PDFì—ì„œ í…ìŠ¤íŠ¸ ì¶”ì¶œ í•¨ìˆ˜
def extract_text_from_pdf(pdf_path, output_path):
    if not os.path.exists(pdf_path):
        print(f"ì˜¤ë¥˜: íŒŒì¼ ê²½ë¡œë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ê²½ë¡œë¥¼ í™•ì¸í•˜ì„¸ìš”: {pdf_path}")
        return

    doc = fitz.open(pdf_path)
    full_text = []

    print(f"PDF í˜ì´ì§€ ìˆ˜: {len(doc)}")

    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        text = page.get_text()
        full_text.append(text)

    doc.close()

    # ì¶”ì¶œëœ í…ìŠ¤íŠ¸ë¥¼ íŒŒì¼ë¡œ ì €ì¥
    with open(output_path, 'w', encoding='utf-8') as f:
        f.write("\n".join(full_text))

    print(f"âœ… í…ìŠ¤íŠ¸ ì¶”ì¶œ ì™„ë£Œ ë° '{output_path}'ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
    return "\n".join(full_text)

# í•¨ìˆ˜ ì‹¤í–‰
bible_full_text = extract_text_from_pdf(pdf_path, extracted_text_path)
```

```python
import re

def chunk_bible_by_verses_v2(text):
    if not text:
        print("ğŸš¨ ì˜¤ë¥˜: ì…ë ¥ í…ìŠ¤íŠ¸ê°€ ë¹„ì–´ ìˆìŠµë‹ˆë‹¤. íŒŒì¼ ë¡œë“œë¥¼ í™•ì¸í•˜ì„¸ìš”.")
        return []

    # ì„±ê²½ í…ìŠ¤íŠ¸ë¥¼ ì ˆ ì •ë³´(ì±…ì´ë¦„ ì¥:ì ˆ)ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë¶„ë¦¬í•˜ëŠ” ì •ê·œì‹
    # ì´ íŒ¨í„´ì€ "í•œê¸€ 2~4ì + ê³µë°± + ìˆ«ì:ìˆ«ì" í˜•ì‹ì„ ì ˆì˜ ì •ë³´ë¡œ ê°€ì •í•©ë‹ˆë‹¤.
    # ì˜ˆì‹œ í¬ë§·: "ì°½ì„¸ê¸° 1:1", "ìš”í•œë³µìŒ 3:16"
    pattern = r'([ê°€-í£]{2,4}\s\d+:\d+)'

    # í…ìŠ¤íŠ¸ë¥¼ íŒ¨í„´(ì ˆ ì •ë³´)ê³¼ íŒ¨í„´ì´ ì•„ë‹Œ ë¶€ë¶„(ì ˆ ë‚´ìš©)ìœ¼ë¡œ ë‚˜ëˆ•ë‹ˆë‹¤.
    parts = re.split(pattern, text)

    verses = []

    # ë¶„í• ëœ ë¦¬ìŠ¤íŠ¸ì—ì„œ 'ì ˆ ì •ë³´'ì™€ 'ì ˆ ë‚´ìš©'ì„ ê²°í•©í•˜ì—¬ ì ˆ ë‹¨ìœ„ ë¦¬ìŠ¤íŠ¸ ìƒì„±
    for i in range(1, len(parts), 2):
        verse_info = parts[i].strip()     # ì˜ˆ: 'ì°½ì„¸ê¸° 1:1'
        verse_content = parts[i+1].strip() # ì˜ˆ: 'íƒœì´ˆì— í•˜ë‚˜ë‹˜ì´ ì²œì§€ë¥¼ ì°½ì¡°í•˜ì‹œë‹ˆë¼.'

        # ìœ íš¨í•œ ì ˆë§Œ ì €ì¥
        if verse_info and verse_content:
            verses.append(f"{verse_info} {verse_content}")

    # 3. 8ê°œì˜ ì ˆì„ í•˜ë‚˜ì˜ ì²­í¬ë¡œ ë¬¶ê¸° (ë¬¸ë§¥ ìœ ì§€ë¥¼ ìœ„í•´ 8ê°œ ì ˆì”© ë¬¶ìŠµë‹ˆë‹¤.)
    final_chunks = []
    verses_per_chunk = 8
    for i in range(0, len(verses), verses_per_chunk):
        # ì²­í¬ê°€ ìƒì„±ë  ë•Œ ì¤„ë°”ê¿ˆ('\n')ìœ¼ë¡œ ì ˆì„ êµ¬ë¶„í•©ë‹ˆë‹¤.
        chunk_content = "\n".join(verses[i:i + verses_per_chunk])
        if chunk_content:
            final_chunks.append(chunk_content)

    print(f"âœ… ì„±ê²½ í…ìŠ¤íŠ¸ê°€ ì ˆ ê¸°ë°˜ìœ¼ë¡œ {len(final_chunks)}ê°œì˜ ì²­í¬ë¡œ ë¶„í• ë˜ì—ˆìŠµë‹ˆë‹¤.")
    print(f"--- ì˜ˆì‹œ ì²­í¬ (ì²« 1ê°œ) ---")
    print(final_chunks[0])

    return final_chunks

# í•¨ìˆ˜ ì‹¤í–‰ ë° ìµœì¢… ì²­í¬ ë³€ìˆ˜ ì €ì¥
bible_chunks_clean = chunk_bible_by_verses_v2(cleaned_bible_text_data)
```
