```python
#실습 2번 숫자 맞추기 게임

import random

a = 0
answer = random.randint(1,1000)
print(answer)

print("1~1000 사이의 정답인 숫자를 입력하면 됩니다.")
for i in range(10,0,-1):
  a = int(input(f"숫자를 입력하세요 기회는 {i}번 : "))
  if answer == a:
    print("정답입니다 축하합니다. 짝짝짝")
    break
  else:
    if a > answer :
      print("작습니다")
    else:
      print("큽니다")
print("수고하셨습니다")
```

```python
#실습 3번 버블정렬

import random

a = []

for i in range(0,20):
  a.append(random.randint(1,100))

length = len(a)-1
temp = 0

for i in range(length,1,-1):
  for j in range(0,length):
    if a[j] >= a[j+1]:
      temp = a[j]
      a[j] = a[j+1]
      a[j+1] = temp

print(a)
```

```python
#실습 4 전화번호 관리 프로그램

data = []

def data_input(a):
  print("===== 전화번호 추가 =====")
  data.append(a)
  print(f'{a["name"]}님의 전화번호가 추가되었습니다.')

def search(s_name):
  print("===== 전화번호 조회 =====")
  length = len(data)
  for i in range(0,length):
    if s_name == data[i]["name"]:
      print(f'이름 : {data[i]["name"]}님의 전화번호 : {data[i]["number"]}')
  print("\n")

def delect():
  print("===== 전체 전화번호 삭제 =====")
  data = []

def all_search():
  print("===== 전체 전화번호 조회 =====")
  length = len(data)
  for i in range(0,length):
    print(f'이름 : {data[i]["name"]} : {data[i]["number"]}')
  print("\n")
while(True):
  print("===== 전화번호 관리 프로그램 =====")
  print("a. 데이터 입력")
  print("b. 조회")
  print("c. 데이터 삭제")
  print("d. 전체 데이터 출력")
  want = input("메뉴를 선택하세요 : ")

  if want == "a":
    name = input("이름을 입력하세요 : ")
    number = input("전화번호를 입력하세요 : ")
    user_data = {"name":name, "number":number}
    data_input(user_data)
  elif want == "b":
    name = input("이름을 입력해주세요 : ")
    search(name)
  elif want == "c":
    delect()
  elif want == "d":
    all_search()
  else:
    break
```

```python
#실습 4-1 전화번호 관리 프로그램 + 로딩 업그레이드

updated_data = []
sub_data = []
loaded_data = []

def data_input(a):
  print("===== 전화번호 추가 =====")
  sub_data.append(a)
  print(f'{a["name"]}님의 전화번호가 추가되었습니다.')

def search(s_name):
  print("===== 전화번호 조회 =====")
  length = len(updated_data)
  for i in range(0,length):
    if s_name == updated_data[i]["name"]:
      print(f'이름 : {updated_data[i]["name"]}님의 전화번호 : {updated_data[i]["number"]}')
  print("\n")

def delect():
  print("===== 전체 전화번호 삭제 =====")
  updated_data = []

def all_search():
  print("===== 전체 전화번호 조회 =====")
  length = len(updated_data)
  for i in range(0,length):
    print(f'이름 : {updated_data[i]["name"]} : {updated_data[i]["number"]}')
  print("\n")

def save_data():
  length = len(sub_data)
  for i in range(0,length):
    updated_data.append(sub_data[i])

def load_data():
  length = len(updated_data)
  for i in range(0,length):
    loaded_data.append(updated_data[i])

while(True):
  print("===== 전화번호 관리 프로그램 =====")
  print("a. 데이터 입력")
  print("b. 조회")
  print("c. 데이터 삭제")
  print("d. 전체 데이터 출력")
  print("e. 데이터 저장")
  print("f. 데이터 로딩")
  print("나머지. 프로그램 종료")
  want = input("메뉴를 선택하세요 : ")

  if want == "a":
    name = input("이름을 입력하세요 : ")
    number = input("전화번호를 입력하세요 : ")
    user_data = {"name":name, "number":number}
    data_input(user_data)
  elif want == "b":
    name = input("이름을 입력해주세요 : ")
    search(name)
  elif want == "c":
    delect()
  elif want == "d":
    all_search()
  elif want == "e":
    save_data()
  elif want == "f":
    load_data()
  else:
    break
```

```python
# 반가산기 구현하기

def and_cal(x1,x2):
  if x1 + x2 == 2:
    return 1
  else:
    return 0

def or_cal(x1,x2):
  if x1 + x2 >= 1:
    return 1
  else:
    return 0
```
